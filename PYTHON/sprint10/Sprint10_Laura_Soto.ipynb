Junta les columnes Fills i No Fills en una sola columna, utilitzant el mètode .apply() i definint una funció que resolgui el problema. La columna nova ha de dir-se "Fills" i prendre els valors "Sí" o "No".

```python
def unir_columnas_fills(row):

    if row["Fills"] == 1:
        return "Sí"
    elif row["No Fills"] == 1:
        return "No"
    else:
        return None

df["Fills"] = df.apply(unir_columnas_fills, axis=1)

df = df.drop(columns=["No Fills"])

df
```

**6.**

Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.

Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python

df["Data Naixement"] = pd.to_datetime(
    df["Dia"].astype(str) + '-' +
    df["Mes"].astype(str) + '-' +
    df["Any"].astype(str),
    format='%d-%m-%Y', dayfirst=True
)

df
```

```python
def calcular_edad(naixement):
    hoy = date.today()
    return (hoy.year - naixement.year) - ((hoy.month, hoy.day) < (naixement.month, naixement.day))

df['Edat'] = df['Data Naixement'].apply(calcular_edad)

df
```

Utilitzant un bucle, exporta en 4 fitxers (format .xlsx o .csv) les dades de cada Grup Professional.

Per exemple: "dades_GrupA.xlsx" , "dades_GrupB.xlsx" ...

Exporta un 5è DataFrame en format .xlsx o .csv que contingui quants treballadors hi ha per cada Grup P

```python
#Dividimos agrupando por la columna que queremos exportar (Grup Professional) para poder iterar

agrupacion = df.groupby("Grup Professional")

#Iteramos sobre los grupos y exportamos los datos de cada uno en un archivo diferente

for nombre_agrupacion, df_agrupacion in agrupacion:
    nombre_archivo = f"dades_{nombre_agrupacion}.xlsx"

    #Exportamos el dataframe
    df_agrupacion.to_excel(
        nombre_archivo,
        index=False,
        sheet_name=f"Hoja{nombre_agrupacion}"
    )

print("Exportación completada.")
```
