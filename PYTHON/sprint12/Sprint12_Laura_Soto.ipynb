**1.** Consulta l’API pública JSONPlaceholder utilitzant el mètode GET per obtenir:

- Llista de publicacions (/posts)

- Llista d’usuaris (/users)

- Llista de tasques (/todos)

```python
import requests
```

Establecemos que la URL de la base de datos que vamos a consultar:

```python
base_URL = "https://jsonplaceholder.typicode.com"
```

Ahora estableceremos qué parte concreta de la URL queremos consultar y guardamos la cadena de texto que recibimos en un json para que Python pueda "entenderlo".
Posteriormente hacemos iteración para que nos muestre el listado de lo que consultamos.

Lista de publicaciones (/posts)

```python
respuesta_posts = requests.get(f"{base_URL}/posts")

posts = respuesta_posts.json()

for post in posts:
    print(post["title"])
```

Lista de usuarios (/users)

```python
respuesta_users = requests.get(f"{base_URL}/users")

users = respuesta_users.json()

for user in users:
     print(f'ID: {user["id"]} | Nombre: {user["name"]} | Username: {user["username"]}')
```

Llista de tasques (/todos)

```python
respuesta_todos = requests.get(f"{base_URL}/todos")

tareas = respuesta_todos.json()

for tarea in tareas:
    print(f'Tarea: {tarea["title"]} | Asignada: {tarea["userId"]} | Completada: {tarea["completed"]}')
```

**2.** Mostra per pantalla:

La quantitat total de cada recurs.

El codi d'estat de cada petició.

Hacemos un print determinando a qué nombre corresponde cada bloque, contamos la cantidad de datos que contiene y verificamos el estado:

```python
print("POSTS")
print("Total:", len(posts))
print("Status code:", respuesta_posts.status_code)
print()

print("USUARIOS")
print("Total:", len(users))
print("Status code:", respuesta_users.status_code)
print()

print("TAREAS")
print("Total:", len(tareas))
print("Status code:", respuesta_todos.status_code)
print()
```

Hay un total de 100 posts, 10 usuarios y 200 tareas. Todos retornan un status code 200, que significa "OK", que la respuesta para la solicitud HTTP ha sido exitosa.

**4.**  Fes una petició POST per crear una nova publicació fictícia. Inclou el títol, el cos del missatge i un userId.

- Mostra la resposta JSON.

- Mostra el codi d'estat.

```python
new_post = {
    "title": "Publicación fictícia de prueba",
    "body": "Este es el cuerpo de la publicación fictícia.",
    "userId": 1
}

respuesta_post = requests.post(f"{base_URL}/posts", json=new_post)

print("Respuesta JSON:")
print(respuesta_post.json())
print("Status code:", respuesta_post.status_code)
```

No guarda los datos realmente pero simula la respuesta de forma correcta. El status code 201 indica que la solicitud se cumplió  y el servidor creó un nuevo recurso.

**5.** Fes una petició PATCH per modificar parcialment una publicació existent.

Mostra la resposta JSON.

Mostra el codi d'estat.

Hacemos modificación del título del primer post: 

```python
patch_data = {
    "title": "Título modificado con PATCH"
}

respuesta_patch = requests.patch(f"{base_URL}/posts/1", json=patch_data)

print("Respuesta JSON:")
print(respuesta_patch.json())
print("Status code:", respuesta_patch.status_code)
```

**6.** Fes una petició DELETE sobre una publicació.

Mostra la resposta JSON.

Mostra el codi d'estat.

```python
respuesta_delete = requests.delete(f"{base_URL}/posts/1")

print("Respuesta JSON:")
print(respuesta_delete.json())
print("Status code:", respuesta_delete.status_code)
```

Vemos que el post 1 aparece vacío
