**1. Calculadora de l'índex de massa corporal**

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def calculo_imc():
    print("Calculadora del Índice de Masa Corporal (IMC)")

    try:
        peso = float(input("Introduce tu peso en kg: "))
        altura = float(input("Introduce tu altura en metros (ej: 1.70):"))

        #Validaciones
        if peso <= 0 or peso > 400:
            print("Error: El peso introducido no es válido.")
            return
        
        if altura < 0 or altura > 3:
            print("Error: La altura introducida no es válida.")
            return
        
        #Cálculo del IMC
        imc = peso/(altura ** 2)

        #Clasificación
        if imc < 18.5:
            categoria = "Peso bajo"
        elif imc < 24.9:
            categoria = "Peso normal"
        elif imc < 29.9:
            categoria = "Sobrepeso"
        else:
            categoria = "Obesidad"
        
        print(f"Tu IMC es: {imc:.2f}")
        print(f"Clasificación: {categoria}")

    except ValueError:
        print("Error: Tienes que introducir valores númericos válidos.")
    
calculo_imc()
```

**2. Convertidor de temperatures**

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
def pedir_temperatura():
    """Solicitamos al usuario que introduzca el texto en ºC y además validamos que se introduzca un valor númerico"""
    try:
        celsius = float(input("Introduce temperatura en ºC: "))
        return celsius
    except ValueError:
        print("Error: Debes escribir un número válido.")
        return None

def validar_rango(celsius):
    """Validamos rango posible"""    
    if celsius < -273.15:
        print("Error: Fuera del rango (-273.15)")
        return False
    return True


def conversion_temperatura(celsius):
    """Realizamos la conversión"""
    fahrenheit = celsius * 1.8 + 32
    kelvin = celsius + 273.15
    return fahrenheit, kelvin

    
def display_resultado(celsius,fahrenheit,kelvin):
    """Muestra los resultados en pantalla"""
    print(f"{celsius}°C es equivalente a:")
    print(f"Fahrenheit: {fahrenheit}°F")
    print(f"Kelvin: {kelvin}K")


def main_conversor():
    temperatura = pedir_temperatura()

    if temperatura is None:
        return #Error ya mostrado
    
    if not validar_rango(temperatura):
        return #Error ya mostrado
    
    fahrenheit, kelvin = conversion_temperatura(temperatura)

    display_resultado(temperatura, fahrenheit, kelvin)

main_conversor()

```

**3. Comptador de paraules d'un text.**

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
import re
from collections import Counter
from pprint import pprint
from statistics import mean

def pedir_texto_al_usuario ():
    texto_usuario = (input ("Escribe el texto que quieras:"))
    return texto_usuario

def programa_cuenta_palabras (texto):
    """Cuenta las palabras introducidas por el usuario y además devuelve la largada media."""    
    palabras_introducidas = re.findall(r'\b\w+\b', texto.lower())

    if not palabras_introducidas:
        print("AVISO: No has introducido palabras en el texto.")
        return
    
    contador_de_palabras = Counter(palabras_introducidas)
    print("\n¿Cuántas palabras tenemos?:")

    pprint(contador_de_palabras)

    media = mean(map(len, palabras_introducidas))

    print("\nEn el texto completo tenemos: ")
    print(f"En total {sum(len(p) for p in palabras_introducidas)} letras")
    print(f"En total {len(palabras_introducidas)} palabras")
    print(f"El largo medio de las palabras es: {media:.2f}")

def main_contador():
    texto = pedir_texto_al_usuario()
    programa_cuenta_palabras(texto)

main_contador()
```

**4. Diccionari invers (amb possibilitat de duplicats)**

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

Fixa’t en el següents exemples per veure l’output esperat.





```python
def cambiar_orden_diccionario(diccionario_original):
    """Función que invierte las claves y valores del diccionario, si el valor aparece más de una vez, se agrupa en una lista."""

    diccionario_invertido = {}
    aviso_duplicados = False

    for clave, valor in diccionario_original.items():
        if valor in diccionario_invertido:
            if isinstance(diccionario_invertido[valor], list):
                diccionario_invertido[valor].append(clave)
            else:
                diccionario_invertido[valor] = [diccionario_invertido [valor], clave]

            if not aviso_duplicados:
                print("AVISO: Hay valores duplicados!")
                aviso_duplicados = True
        else:
            diccionario_invertido [valor] = clave
    
    return diccionario_invertido

def main_diccionario_inverso():
    """Función que contiene el diccionario con el cual vamos a trabajar"""
    diccionario_original = {
        "nombre 1" : "Etna", 
        "nombre 2" : "Laura", 
        "nombre 3" : "Enzo", 
        "nombre 4" : "Laura"}
    
    resultado_final = cambiar_orden_diccionario(diccionario_original)

    print("\nDiccionario invertido final:")
    print(resultado_final)

main_diccionario_inverso()
```

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:



```python
import re
from collections import Counter
from pprint import pprint

def texto_externo(nombre_archivo):
    """ Lee el contenido del texto y lo retorna como string"""
    with open(nombre_archivo, "r", encoding='utf-8', errors="replace") as archivo:
        return archivo.read()

def cuenta_palabras_por_letras(texto):
    """Extrae las palabras del texto y las agrupa por su letra inicial"""
    palabras = re.findall(r'\b\w+\b', texto.lower())
    lista_de_palabras = {}

    for palabra in palabras:
        inicial = palabra [0] #primera letra

        #Si la letra no está en el diccionario, la contamos
        if inicial not in lista_de_palabras:
            lista_de_palabras[inicial] = Counter()
        
        lista_de_palabras[inicial][palabra] += 1

    lista_ordenada = {
         letra: dict(sorted(contador.items()))
         for letra, contador in lista_de_palabras.items()
    }

    return lista_ordenada

def main_contador_texto():
        nombre_archivo = r"C:\Users\rubit\Desktop\Formaciones\IT ACADEMY\Nivel II\PYTHON\Sprint9\BBDD\tu me quieres blanca.txt"

        texto = texto_externo(nombre_archivo)

        resultado= cuenta_palabras_por_letras(texto)

        pprint(resultado)

main_contador_texto()
        
```

**2. Conversió de tipus de dades.**

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:



```python
def lista_de_datos():
    lista = ["39", "2.3", ("4.1", "quince"), ["mil", "2", ("hola_que_tal",)], "1e3", "4", "x", ["34*3/7"]]
    return lista

def iterable_no_string(elemento):
    return isinstance(elemento,(list,tuple)) 

def dividir_listas(datos):
    lista_flotantes = []
    lista_no_flotantes = []

    def navega_lista_y_tupla(elemento):
        if iterable_no_string(elemento):
            for sub in elemento:
                navega_lista_y_tupla(sub)
            return
        
        try:
            convertirlo = float(elemento)
            lista_flotantes.append(convertirlo)
        
        except (ValueError, TypeError):
            lista_no_flotantes.append(elemento)
    
    navega_lista_y_tupla(datos)
    return lista_flotantes, lista_no_flotantes

def main_revisar_datos_lista():
    datos = lista_de_datos()
    flotantes, no_flotantes = dividir_listas(datos)

    print("Elementos convertidos a float:")
    print(flotantes)

    print("\nElementos que no se han podido convertir:")
    print(no_flotantes)

main_revisar_datos_lista()
```

Explora el funcionament del mòdul random de la llibreria numpy. 🔗 Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

longitud (int): Longitud de la contrasenya

majuscules (bool = True): Si hi ha d’aparèixer majúscules

minuscules (bool = True): Si hi ha d’aparèixer minúscules

Així doncs, si executem la funció de la següent manera:

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

```python
#importamos las librerías
import numpy as np
import string
import pyperclip
import random

def crear_contrasenya(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):
    """Generamos una contraseña automática con los caracteres permitidos"""
    
    #Construimos el conjunto de caracteres permitidos
    caracteres = ""
    grupos = []
    obligatorios = []

    if mayusculas: # ABC...
        grupos.append(string.ascii_uppercase)
        obligatorios.append(random.choice(string.ascii_uppercase))
    if minusculas: # abc...
        grupos.append(string.ascii_lowercase)
        obligatorios.append(random.choice(string.ascii_lowercase))
    if numeros: # 0123456789
        grupos.append(string.digits)
        obligatorios.append(random.choice(string.digits))          
    if signos: # símbolos: , - $ ?
        grupos.append(string.punctuation)
        obligatorios.append(random.choice(string.punctuation))      


    # Construimos el conjunto total
    caracteres = "".join(grupos)

    # Rellenamos el resto de la contraseña
    resto = [random.choice(caracteres) for _ in range(longitud - len(obligatorios))]

    # Combinar obligatorios + aleatorios
    contrasenya_lista = obligatorios + resto

    # Mezclar para que no queden fijos al inicio
    random.shuffle(contrasenya_lista)


    # Generamos contraseña con numpy
    contrasenya = "".join(contrasenya_lista)

    #EXTRA copiar contraseña con pyperclip
    pyperclip.copy(contrasenya)

    return contrasenya

crear_contrasenya(10, True, True, True, True)
```

**2. Processament de dades simple**

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l’equip més golejador.
El nom de l’equip més golejat
La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)

```python
#Importamos la librería pandas
import pandas as pd

def cargar_archivo(archivo_importado, columnas_nuevas):
    return pd.read_csv(archivo_importado, sep="\t", header=None, names=columnas_nuevas)

def separar_goles(df):
    """Sepero los números de la columna resultados y convierto en enteros"""
    df[["Goles Local", "Goles Visitante"]] = df["Resultado"].str.split("-", expand=True).astype(int)
    return df

def calculo_estadisticas(df):
    # Goles a favor / en contra
    goles_a_favor = (df.groupby("Equipo Local")["Goles Local"].sum().add(df.groupby("Equipo Visitante")["Goles Visitante"].sum(),           fill_value=0))
    
    goles_en_contra = (df.groupby("Equipo Local")["Goles Visitante"].sum().add(df.groupby("Equipo Visitante")["Goles Local"].sum(), fill_value=0))

    # Resultados del partido
    df["Local_ganados"], df["Visitante_ganados"], df["Empate"] = df["Goles Local"]>df["Goles Visitante"], df["Goles Visitante"]>df["Goles Local"], df["Goles Local"]==df["Goles Visitante"]

    # Ganados / Empatados
    partidos_ganados = (df.groupby("Equipo Local")["Local_ganados"].sum().add(df.groupby("Equipo Visitante")["Visitante_ganados"].sum(), fill_value=0))
    
    partidos_empatados = (df.groupby("Equipo Local")["Empate"].sum().add(df.groupby("Equipo Visitante")["Empate"].sum(), fill_value=0))


    # Clasificación general
    clasificacion = pd.DataFrame({
        "Goles Totales": goles_a_favor,
        "Puntos": partidos_ganados*3 + partidos_empatados
    }).sort_values("Puntos", ascending=False)

     # Equipo más goleador y más goleado
    equipo_mas_goleador = goles_a_favor.idxmax()  # Devuelve el equipo con más goles
    equipo_mas_goleado = goles_en_contra.idxmax()  # Devuelve el equipo con más goles en contra

    return clasificacion, equipo_mas_goleador, equipo_mas_goleado


def main_tabla_final():
    columnas = ["Equipo Local", "Resultado", "Equipo Visitante"]

    archivo_importado = (r'C:\Users\rubit\Desktop\Formaciones\IT ACADEMY\Nivel II\PYTHON\Sprint9\BBDD\historic partits.txt')

    df = cargar_archivo(archivo_importado, columnas)
    df = separar_goles(df)
    
    clasificacion, equipo_mas_goleador, equipo_mas_goleado = calculo_estadisticas(df)

    print("Equipo más goleador:", equipo_mas_goleador)
    print("Equipo más goleado:", equipo_mas_goleado)

    return clasificacion
    
tabla =  main_tabla_final()

tabla
```
